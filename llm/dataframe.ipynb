```python
import pandas as pd
#I will read all the csvs 
df_agrarias = pd.read_csv("results/GERAL/ciencias_agrarias.csv", sep="|")
df_agrarias = df_agrarias.assign(Área="Ciências Agrárias")

df_biologias = pd.read_csv("results/GERAL/biologias.csv", sep="|")
df_biologias = df_biologias.assign(Área="Biológicas")

df_engenharia = pd.read_csv("results/GERAL/engenharia.csv", sep="|")
df_engenharia = df_engenharia.assign(Área="Engenharia")

df_humanas = pd.read_csv("results/GERAL/humanas.csv", sep="|")
df_humanas = df_humanas.assign(Área="Humanas")

df_saude = pd.read_csv("results/GERAL/saude.csv", sep="|")
df_saude = df_saude.assign(Área="Saúde")

df_social = pd.read_csv("results/GERAL/social.csv", sep="|")
df_social = df_social.assign(Área="Social")
#now merge the dataframes
df = pd.concat([df_agrarias, df_biologias, df_engenharia, df_humanas, df_saude, df_social], ignore_index=True)
df = df.drop(columns=["Tipo de Acesso", "Idioma"])
# df.to_csv("results/GERAL/ufrgsv1.csv", sep="|", index=False)
# print(df.size)
# df_ufrgsv1 = pd.read_csv("results/GERAL/ufrgsv1.csv", sep="|")
# print(df_ufrgsv1.size)
```
